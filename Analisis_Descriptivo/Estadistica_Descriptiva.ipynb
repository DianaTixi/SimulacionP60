# Analisis Datos Fifa

```python
#Librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos_fifa= pd.read_csv("https://github.com/DianaTixi/SimulacionP60/raw/main/Datos-Fifa/fifa_datos.csv")
datos_fifa=datos_fifa.drop(columns = ['Unnamed: 0'])
datos_fifa.head()
```

```python
datos_fifa.info()
```

```python
datos_fifa.describe()
```

```python
datos_fifa.corr(method ='pearson')
```

```python
sns.heatmap(datos_fifa.corr())
```

```python
data2_fifa= datos_fifa.loc[:,['Age', 'Overall','Potential', 'Skill Moves','International Reputation']]
sns.heatmap(data2_fifa.corr())
```
