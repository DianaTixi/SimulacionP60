1. Con los datos de Fifa, organizar a los jugadores de acuerdo al peso en las siguientes escalas y generar un cuadro tipo PIE

  * Debajo 125 Lbs.
  * 125-150
  * 150-175
  * 175 o superior

```python
#Librerias 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io
from google.colab import files
import plotly.graph_objects as go
```

```python
#Abrir CSV
url = 'sample_data/fifa_datos.csv'
df = pd.read_csv(url, delimiter=',')
df
```

```python
# Generar un grafico de cual es su pie diestro
df['Weight']= df['Weight'].replace({'lbs':''}, regex=True) 
df= df.dropna(subset=['Weight'])
df['Weight'] = df['Weight'].astype(int)

inferior = df.loc[df['Weight'] < 125].count()[0]
medio1 = df.loc[(df['Weight'] >=125) & (df['Weight'] <150)].count()[0]
medio2 = df.loc[(df['Weight'] >=150) & (df['Weight'] <175)].count()[0]
superior = df.loc[df['Weight'] >=175].count()[0]

plt.figure()

etiquetas = ['Debajo 125 lbs','125 - 150 lbs', '150 - 175 lbs', 'Superior 175 lbs']
colores = ['#abcdef', '#aabbcc']
plt.pie([inferior, medio1, medio2, superior], labels= etiquetas, colors=colores, autopct='%.2f %%')
plt.title('Porcentaje de Jugadores segun su peso')
plt.show()
```
