```python
# Cuadrados medios: Xo=74731897457, D=7
xo = 74731897457
digitos = 7
iteraciones =100
xn = xo
lista1=[]
#print("It. \t Xn \t Ui \t Rn")
for i in range(iteraciones):
  xnn = xn**2
  txnn = str(xnn).zfill(8)
  tam = len(txnn)
  ui = int(txnn[int(tam/2-digitos/2):int(tam/2+digitos/2)])
  rn = ui / (int('9'*digitos)+1)
  lista1.append(rn)
  xn = ui
print("Lista1= " , lista1)
  
  
```

```python
# Congruencia lineal: a=74731897457, b=37747318974, Xo=7, M=19
xo =7
a = 74731897457
b = 37747318974
m = 19
iteraciones = 100
xn = xo
lista2=[]
#print("It. \t Xn \t Ui")
for i in range(iteraciones):
  xnn = (a*xn+b) % m
  ui = xnn/m
  lista2.append(ui)
  xn = xnn
print("Lista2=  ", lista2)
```
