###### **Problematica**: La distribucion del coste de un siniestro de una cartera de seguros, sigue ana distribucion de PARETO de parametros 


*   α= 2
*   β= 300



###### **Simulacion**

```python
import numpy as np
import random
import matplotlib.pyplot as plt

a=2 
b=300
u=[]
resultado=[]

for i in range(100):
  u.append(random.random())
  
for i in range(len(u)):
  formula= b/((u[i])**(1/a))
  resultado.append(formula)
```

```python
print(resultado)
```

##### **Graficas**

```python
count, x, _ = plt.hist(resultado, 100, density=True) 
densidad = a*b**a / x**(a+1) 
plt.plot(x, max(count)*densidad/max(densidad), linewidth=2, color='yellow')
plt.title('Función de Densidad')
plt.show()
```

```python
count, x, _ = plt.hist(resultado, 100, density=True) 
acumulada = 1 - (b/x)**a 
plt.plot(x, max(count)*acumulada/max(acumulada), linewidth=2, color='yellow') 
plt.title('Función Acumulada')
plt.show()
```
